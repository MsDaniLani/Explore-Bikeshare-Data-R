```r
# Function to view column names, number of columns, and number of rows.

view_data_info <- function(data) {
  #'''
#' This function takes a dataframe as input and prints the column names, number of columns, and number of rows.
#'
#' parameter - data - A dataframe.
#' return value - None - prints the information to the console
#'''
  cat("Column Names:\n")
  print(colnames(data))
  cat("\nNumber of Columns: ", ncol(data))
  cat("\nNumber of Rows: ", nrow(data), "\n")  
}
```

### Cleaning and Transforming Data

```r
# Add missing columns to the Washington dataframe so it matches Chicago and New York in order to combine.
wash['Gender'] <- NA
wash['Birth.Year'] <- NA

# Run the function view_data_info to view information for Washington was added correctly.
view_data_info(wash)
```

```r
# Add a 'city' column to each dataframe so when combined the city is retained.
ny['City'] <- 'New York City'
wash['City'] <- 'Washington'
chi['City'] <- 'Chicago'
```

```r
# View column names for all dataframe to was added correctly
cat('Chicago Column Names')
colnames(chi)

cat('New York Column Names')
colnames(ny)

cat('Washington Column Names')
colnames(wash)
```

```r
# Combine the dataframes using rbind()
all_cities <- rbind(ny, wash, chi)

# Run the function view_data_info to view information for combined dataframe
view_data_info(all_cities)
```

### Question 1

**Popular times of travel**
- What is the most common month?
- What is the most common day of week?
- What is the most common hour of day?

```r
# Determine the format of the Start.Time column
strptime(all_cities$Start.Time[1], "%Y-%m-%d %H:%M:%S")

# Convert Start.Time column to datetime format using the determined format
all_cities$Start.Time <- as.POSIXct(all_cities$Start.Time, format = "%Y-%m-%d %H:%M:%S")

# Extract the month, day of week, and hour of day
all_cities$Month <- format(all_cities$Start.Time, "%m")
all_cities$DayOfWeek <- format(all_cities$Start.Time, "%A")
all_cities$HourOfDay <- format(all_cities$Start.Time, "%H")

# Calculate the frequency of each value and find the most common
most_common_month <- names(which.max(table(all_cities$Month)))
most_common_day <- names(which.max(table(all_cities$DayOfWeek)))
most_common_hour <- names(which.max(table(all_cities$HourOfDay)))

# Print the result
cat("The most common month is", most_common_month, "\n")
cat("The most common day of week is", most_common_day, "\n")
cat("The most common hour of day is", most_common_hour, "\n")
```

```r
# Create a bar chart of the frequency of usage each month
library(ggplot2)
ggplot(all_cities, aes(x = Month)) +
  geom_bar(fill = "pink") +
  labs(title = "Frequency of Trips by Month", x = "Month", y = "Frequency")

# Create a bar chart of the frequency of usage each day of week
ggplot(all_cities, aes(x = DayOfWeek)) +
  geom_bar(fill = "turquoise") +
  labs(title = "Frequency of Trips by Day of Week", x = "Day of Week", y = "Frequency")

# Create a bar chart of the frequency of usage each hour of day
ggplot(all_cities, aes(x = HourOfDay)) +
  geom_bar(fill = "light green") +
  labs(title = "Frequency of Trips by Hour of Day", x = "Hour of Day", y = "Frequency")
```

### Summary of Question 1

##### The most common month is 06 which is June. The data used is based on northern states. During the month of June, the weather is generally warm and pleasant, which may contribute to the high frequency of bike trips.

##### The most common day of the week is Wednesday, suggesting that people may be using bikeshare for commuting or other regular activities during the workweek.

##### The most common hour of day is 08, which is likely a peak commuting time. This information can be useful for bike sharing companies to optimize their operations and ensure that there are enough bikes available during high-demand times.

### Question 2


**Popular stations and trip**
- What is the most common start station?
- What is the most common end station?
- What is the most common trip from start to end?

```r
# Calculate the most common start station.
most_common_start_station <- names(which.max(table(all_cities$Start.Station)))

# Calculate the most common end station.
most_common_end_station <- names(which.max(table(all_cities$End.Station)))

# Calculate the most common trip.
most_common_trip <- names(which.max(table(paste(all_cities$Start.Station, all_cities$End.Station, sep = " to "))))

# print the result
cat("The most common start station is", most_common_start_station, "\n")
cat("The most common end station is", most_common_end_station, "\n")
cat("The most common trip is from", most_common_trip, "\n")
```

### Summary of Question 2
##### The most common start station is Columbus Circle / Union Station 
##### The most common end station is Columbus Circle / Union Station 
##### The most common trip is from Jefferson Dr & 14th St SW to Jefferson Dr & 14th St SW 

### Question 3

**Trip duration**

- What is the total travel time for users in different cities for each user type?
- What is the average travel time for users in different cities?

```r
# calculate the average trip duration for each user type in each city
avg_duration <- aggregate(Trip.Duration ~ City + User.Type, data = all_cities, FUN = mean)

# print the result
avg_duration
```

```r
# calculate the average travel time by city
avg_travel_time <- aggregate(Trip.Duration ~ City, data = all_cities, FUN = mean)

# create a vector of colors for each city
city_colors <- c("turquoise", "light green", "pink")

# create a bar plot of total travel time by city with color
barplot(avg_travel_time$Trip.Duration, main = "Total Travel Time by City", xlab = "City", ylab = "Total Travel Time (seconds)", col = city_colors)

# create a bar plot of average travel time by city with color
barplot(avg_travel_time$Trip.Duration, main = "Average Travel Time by City", xlab = "City", ylab = "Average Travel Time (seconds)", col = city_colors)
```

### Summary of Question 3
##### This gives an idea of how efficient bike share services are in each city.

##### For example, bike share rides in Washington DC take 1,233 seconds on average (about 20 minutes). This is longer than in Chicago (937 seconds or 15 minutes) and New York City (904 seconds or 15 minutes). So bike share services in Washington DC may be less efficient.

##### However, we need to look at other details to understand bike share usage accurately and make useful conclusions about how good the services are in these cities.

### Question 4

**User info**

- What are the counts of each user type?
- What are the counts of each gender (only available for NYC and Chicago)?
- What are the earliest, most recent, most common year of birth (only available for NYC and Chicago)?

```r
# Calculate total of each user type.
user_types <- table(all_cities$User.Type)

# Calculate total of each gender.
gender_counts <- table(all_cities$Gender)

# Calculate the earliest, most recent, and most common year of birth.
if ("Birth.Year" %in% colnames(ny) && "Birth.Year" %in% colnames(chi)) {
  birth_years <- subset(all_cities, select = c(Birth.Year))$Birth.Year
  earliest_birth_year <- min(na.omit(birth_years))
  most_recent_birth_year <- max(na.omit(birth_years))
  most_common_birth_year <- names(which.max(table(na.omit(birth_years))))
}

# Print the results
cat("Counts of each user type:\n")
print(user_types)

if (exists("gender_counts")) {
  cat("\nCounts of each gender:\n")
  print(gender_counts)
}

if (exists("earliest_birth_year")) {
  cat("\nEarliest birth year: ", earliest_birth_year, "\n")
  cat("Most recent birth year: ", most_recent_birth_year, "\n")
  cat("Most common birth year: ", most_common_birth_year, "\n")
}
```

```r
# Calculate the total of each gender.
gender_counts <- table(all_cities$Gender)

# Create a vector of colors for each user type.
user_type_colors <- c("pink", "light green", "turquoise")

# Create a pie chart of user type with "Unknown" labeled.
user_type_labels <- ifelse(gender_counts == 0, "Unknown", names(gender_counts))

# Plot the pie chart with labels and number amounts.
pie(gender_counts, labels = paste(user_type_labels, "(", gender_counts, ")"), col = user_type_colors)

# Add a legend to the chart.
legend("topright", legend = c("Unknown", "Female", "Male"), fill = user_type_colors)
```

### Summary of Question 4

##### The user types in the dataframe are divided into "Subscribers" and "Customers." There are a total of 121,576 subscribers and 30,754 customers. Additionally, there are 121 entries with an unknown user type.

##### When analyzing gender, it is observed that the primary users are males, with a count of 42,360. The count for females is 13,882, and there are 7,158 entries with an unknown gender.

##### The birth year information reveals additional details:
- The earliest birth year recorded in the dataframe is 1885.
- The most recent birth year recorded is 2002.
- The most common birth year among the users is 1989.

**The Washington dataframe did not contain `Gender` or `Birth.Year`, those columns were added and filled NA in order to combine the dataframe.**
